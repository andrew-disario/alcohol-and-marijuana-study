```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/drug-use-by-age/drug-use-by-age.csv')
data
```

```python
data.columns = [i.replace('-', '_') for i in data.columns]
```

```python
plt.plot(data.age, (data.alcohol_use * data.alcohol_frequency) / 1000)
plt.plot(data.age, (data.marijuana_use * data.marijuana_frequency) / 1000)
plt.legend(['Alcohol', 'Marijuana'])
plt.xlabel('Age')
plt.ylabel('Usage')

plt.show()
```
