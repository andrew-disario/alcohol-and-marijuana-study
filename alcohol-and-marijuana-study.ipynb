```python
# import libraries
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
# path in data
data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/drug-use-by-age/drug-use-by-age.csv')

data
```

```python
# replace "-" with "_" in column names
data.columns = [i.replace('-', '_') for i in data.columns]
```

```python
# plot data
plt.plot(data.age, (data.alcohol_use * data.alcohol_frequency) / 1000, color = 'magenta')
plt.plot(data.age, (data.marijuana_use * data.marijuana_frequency) / 1000, color = 'cyan')
plt.xticks(rotation = 90)
plt.legend(['Alcohol', 'Marijuana'])
plt.xlabel('Age')
plt.ylabel('Usage')

plt.show()
```
